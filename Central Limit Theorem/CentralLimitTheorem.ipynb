### Central Limit Theorem Application:
This notebook demonstrates the Central Limit Theorem (CLT) using a population of random data points. The CLT states that the distribution of the sample means approximates a normal distribution, regardless of the population's distribution, given a sufficiently large sample size


```python
# importing the libraries 
import numpy as np
import matplotlib.pyplot as plt
```

1. **Setting the Seed**: We set the random seed to `42` to ensure reproducibility.
2. **Generating Population Data**: We generate `3000` random data points from a gamma distribution with shape parameter `1` and scale parameter `100`.
3. **Plotting the Population Data**: We plot a histogram of the population data to visualize its distribution.
4. **Calculating the Mean**: We calculate and display the mean of the population data.

```python
# creating the population randomized 3000 random data points and calculating its mean
np.random.seed(42)
population_data = np.random.gamma(1, 100, 3000)
plt.hist(population_data)
population_data.mean()
```

* From ***popultion_data*** we will loop random 10000 samples consist of 3 random numbers 
* we calculate the mean of each sample then append it in ***"sample_size_3_means"*** list
* then I will plot a histogram

```python
# creating 10000 samples of 3 random numbers and calculating its mean 
sample_size_3_means = []
for _ in range(10000):
    sample_3= np.random.choice(population_data, 3)
    sample_size_3_means.append(sample_3.mean())
plt.hist(sample_size_3_means)
```

```python
np.array(sample_size_3_means).mean()
```

```python
sample_size_100_means = []
for _ in range(10000):
    sample_100= np.random.choice(population_data, 100)
    sample_size_100_means.append(sample_100.mean())
plt.hist(sample_size_100_means,  )
```

```python
np.array(sample_size_100_means).mean()
```
